```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
df = pd.read_csv('modstudent.csv')
X = df[[ 'gender', 'caste', 'mathematics_marks','english_marks', 'science_marks', 'science_teacher','languages_teacher', 'guardian', 'internet', 'school_id','total_students', 'total_toilets', 'establishment_year']]
y=df['continue_drop']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
print([y.value_counts(),y_train.value_counts(),y_test.value_counts()])
```

```python
y.value_counts()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
gx = []
gy = []
for i in range(1,5):
    gx.append(i)
    clf = RandomForestClassifier(n_estimators=i)
    clf = clf.fit(X_train,y_train)
    clf.predict(X_test)
    gy.append(clf.score(X_test,y_test))
plt.figure(figsize=(8,8))    
plt.plot(gx,gy)
plt.title("Estimators vs Accuracy")
```
