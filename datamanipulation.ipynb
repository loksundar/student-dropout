```python
def transf(dat):
    atr=[]
    for x in dat['continue_drop']:
        if x=="continue":
            atr.append(0)
        else:
            atr.append(1)
    dat['continue_drop']=atr
    gen=[]
    for x in dat['gender'] :
        if x=="M" :
            gen.append(1)
        else :
            gen.append(2)
    dat['gender']=gen
    inter=[]
    for x in dat['internet'] :
        if x==0 :
            inter.append(0)
        elif x==1 :
            inter.append(1)
    dat['internet']=inter
    gua=[]
    for x in dat['guardian']:
        if x=="mother" :
            gua.append(2)
        elif x=="father" :
            gua.append(1)
        else:
            gua.append(0)
    dat['guardian']=gua
    caste=[]
    for x in dat['caste']:
        if x=="OC" :
            caste.append(1)
        elif x=="BC" :
            caste.append(2)
        elif x=="SC" :
            caste.append(3)
        elif x=="ST":
            caste.append(4)
    dat['caste']=caste
    return dat
```
