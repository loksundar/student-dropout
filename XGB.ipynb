```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
df = pd.read_csv('modstudent.csv')
X = df[[ 'gender', 'caste', 'mathematics_marks','english_marks', 'science_marks', 'science_teacher','languages_teacher', 'guardian', 'internet', 'school_id','total_students', 'total_toilets', 'establishment_year']]
y=df['continue_drop']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from xgboost import XGBClassifier
```

```python
gx = []
gy = []
for i in range(1,10):
    gx.append(i)
    xb = XGBClassifier(random_state=0, n_jobs=-1, learning_rate=0.1, n_estimators=5, max_depth=i)
    xb.fit(X_train,y_train)
    gy.append(100*xb.score(X_test,y_test))
plt.figure(figsize=(8,8))  
plt.plot(gx,gy)
plt.title("DEPTH vs ACCURACY")
```

```python
gx = []
gy = []
for i in range(1,25):
    gx.append(i)
    xb = XGBClassifier(random_state=0, n_jobs=-1, learning_rate=0.1, n_estimators=i, max_depth=5)
    xb.fit(X_train,y_train)
    gy.append(100*xb.score(X_test,y_test))
plt.figure(figsize=(8,8))  
plt.plot(gx,gy)
plt.title("ESTIMATORS vs ACCURACY WITH DEPTH =5")
ax = plt
plt.show()
```
